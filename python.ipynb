```python
def fib(n):
  """This functions takes the number of elements you want to print in
      fibonacci series"""
  if (n<=0):
   return -1
  elif (n==1):
   return [0]
  elif (n==2):
   return [0,1]
  else:
   seq = fib(n-1)
   seq.append(seq[-1]+seq[-2])
   return seq

```

```python
class Student():
  def __init__(self,name,id):
    self.name = name
    self.id = id

  @property
  def get(self):
    print(f"{self.name} has student id {self.id}")



```

```python
class Library(Student):
  def __init__(self,name,id,books=[]):
   self.books = books
   self.name = name
   self.id = id

  @property
  def get(self):
    print(f"{self.name} has issued {len(self.books)} books:")
    if (len(self.books)>0):
      for i in self.books:
        print(i)

  def set_book(self,book):
    for i in book:
     self.books.append(i)
    print("Book List updated. ")

```

```python
s1 = Student("Mayank",200192)
s1.get

```

```python
l1 = Library("Mayank",200192,['Book 1'])
```

```python
l1.get
```
